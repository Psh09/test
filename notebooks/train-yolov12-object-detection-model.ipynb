```python
!pip install ultralytics

from ultralytics import YOLO

# Load a pretrained YOLOv8n model (as YOLOv12 is not a standard model name, using YOLOv8 as a common and performant alternative)
# Assuming YOLOv12 was a typo or refers to a specific version/configuration based on YOLO architecture.
# If a specific YOLOv12 model is intended and available, the model loading line should be adjusted accordingly.
model = YOLO("yolov8n.pt")  # Load a pretrained nano model

# Train the model
# The data path should point to the data.yaml file created by Roboflow download, which contains the paths to train and valid directories and class information.
# Assuming the data.yaml file is located at the dataset.location path.
data_yaml_path = os.path.join(dataset_path, "data.yaml")

results = model.train(data=data_yaml_path, epochs=50, imgsz=640, name="yolov8n_custom")
```
